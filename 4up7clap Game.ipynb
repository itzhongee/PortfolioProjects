```python
# create a programme that prints the numbers to say during the 4up7clap game
# players have to take turns to say numbers from 1 to 100 
# but they have to say 'up' for numbers that contain 4 or are multiples of 4 and clap for numbers that contain 7 or are multiples of 7
# and say 'up' and clap at the same time for numbers that contain both 4 and 7 or are multiples of both 4 and 7

# function to determine if a digit d1 or d2 is in the number x
def isDigitPresent(x,d1):
    # Break loop if d is present as digit
    while (x > 0):
        if (x % 10 == d1):
            break
        x = x / 10
    # If loop broke
    return (x > 0)

def isDigitPresent(x,d2):
    # Break loop if d is present as digit
    while (x > 0):
        if (x % 10 == d2):
            break
        x = x / 10
    # If loop broke
    return (x > 0)
```

```python
# Game Part 1 (for loop)
# d1 is number to say up
n0 = 1
n1 = 50
d1 = 4

# function to display the values
def printNumbers(n1, d1):
    # Iterate over all numbers to determine whether to say up or just say the number
    for x in range(n0, n1+1):
        if (x == d1 or isDigitPresent(x, d1)):
            print('say up')
        elif x%d1==0:
            print('say up')
        else:
            print(x)

printNumbers(n1, d1)
```

```python
# Game Part 1 (list comprehension)
# d1 is number to say up
n0 = 1
n1 = 50
d1 = 4

game_1 = ['say up' if x==d1 or isDigitPresent(x,d1) or x%d1==0 else x for x in range(n0,n1+1)]
game_1
```

```python
# Game Part 2 (for loop)
# d2 is number to clap
n0 = 51
n1 = 100
d2 = 7

# function to display the values
def printNumbers(n1, d2):
    # Iterate over all numbers to determine whether to clap or just say the number
    for x in range(n0, n1+1):
        if (x == d2 or isDigitPresent(x, d2)):
            print('clap')
        elif x%d2==0:
            print('clap')
        else:
            print(x)

printNumbers(n1, d2)
```

```python
# Game Part 2 (list comprehension)
# d2 is number to clap
n0 = 51
n1 = 100
d2 = 7

game_2 = ['clap' if x==d2 or isDigitPresent(x,d2) or x%d2==0 else x for x in range(n0,n1+1)]
game_2
```

```python
# Game Part 3 (for loop)
# d1 is number to say up and d2 is number to clap
n0 = 1
n1 = 50
d1 = 4
d2 = 7

# function to display the values
def printNumbers(n1, d1, d2):
    # Iterate over all numbers to determine whether to say up, clap, do both or just say the number
    for x in range(n0, n1+1):
        if (x == d1 or isDigitPresent(x, d1)) and (x == d2 or isDigitPresent(x, d2)):
            print('say up and clap')
        elif x%d1==0 and x%d2==0:
            print('say up and clap')
        elif (x == d1 or isDigitPresent(x, d1)):
            print('say up')
        elif x%d1==0:
            print('say up')
        elif (x == d2 or isDigitPresent(x, d2)):
            print('clap')
        elif x%d2==0:
            print('clap')
        else:
            print(x)

printNumbers(n1, d1, d2)
```
