```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

covid = pd.read_csv('D:/SMU/Data Analytics/Datasets/covid_data.csv')
covid.head()
```

```python
covid.isnull().sum()
```

```python
covid['total_cases'].fillna(0,inplace=True)
covid['total_deaths'].fillna(0,inplace=True)
covid['new_cases'].fillna(0,inplace=True)
covid['new_cases_smoothed'].fillna(0,inplace=True)
covid['new_deaths'].fillna(0,inplace=True)
covid['new_deaths_smoothed'].fillna(0,inplace=True)
covid.isnull().sum()
```

```python
covid.head()
```

```python
#filter data based on specific countries
covid_sg = covid[covid['location']=='Singapore']
covid_us = covid[covid['location']=='United States']
covid_europe = covid[covid['continent']=='Europe']
covid_asia = covid[covid['continent']=='Asia']
```

```python
#filter data based on cases before and after Singapore began its vaccination program
covid_sg_vacc0 = covid_sg[covid_sg['date'] < '2021-01-11']
covid_sg_vacc1 = covid_sg[covid_sg['date'] >= '2021-01-11']
```

```python
from scipy.stats import ttest_ind
data1 = covid_sg_vacc0['new_cases']
data2 = covid_sg_vacc1['new_cases']
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('New average cases are not statistically lower')
else:
    print('New average cases are statistically lower')
```

```python
from scipy.stats import ttest_ind
data1 = covid_sg_vacc0['new_deaths']
data2 = covid_sg_vacc1['new_deaths']
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('New average deaths are not statistically lower')
else:
    print('New average deaths are statistically lower')
```
