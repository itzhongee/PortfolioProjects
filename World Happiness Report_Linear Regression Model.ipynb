```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt

df = pd.read_csv('D:/SMU/Data Analytics/Datasets/world_happiness_report.csv')

df.isnull().sum()

X = df[['Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']].values
y = df[['Ladder score']]
print(X,y)

#Linear regression model algorithm
l_reg = linear_model.LinearRegression()

plt.scatter(X.T[5], y)
plt.show()

#create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = l_reg.fit(X_train, y_train)
predictions = model.predict(X_test)

print("predictions: ", predictions)
print("actual: ", y_test)
print("R^2 value: ", l_reg.score(X,y))
print("coefficient: ", l_reg.coef_)
print("intercept: ", l_reg.intercept_)


#Regression: Score = -2.105 + 0.305*logGDP + 2.272*SS + 0.027*LE + 1.842*Freedom + 0.272*Generosity -0.402*Corruption
#R^2 = 0.752
```
