```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

au = pd.read_csv('D:/SMU/Data Analytics/Datasets/AmongUs.csv')

au.isnull().sum()

imp_au = au[au['Team']==1]
crew_au = au[au['Team']==0]

#if imposter
X = imp_au.drop(columns = ['Outcome'], axis=1)
y = imp_au['Outcome']

#create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model_imp = LogisticRegression()

model_imp.fit(X_train, y_train)

predictions = model_imp.predict(X_test)

# predict the target on the train dataset
predict_train = model_imp.predict(X_train)
print('predictions on train dataset are: ',predict_train) 

# Accuray Score on train dataset
accuracy_train = accuracy_score(y_train,predict_train)
print('accuracy of train dataset is: ', accuracy_train)

# predict the target on the test dataset
predict_test = model_imp.predict(X_test)
print('predictions on test dataset are: ',predict_test) 

# Accuracy Score on test dataset
accuracy_test = accuracy_score(y_test,predict_test)
print('accuracy of test dataset is: ', accuracy_test)
```
