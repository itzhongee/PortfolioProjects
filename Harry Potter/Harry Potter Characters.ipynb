```python
import json
import pandas as pd
```

```python
character = pd.read_csv('Characters.csv', sep=";")
character
```

```python
character_short = pd.read_csv('shortversioncharacters.csv')
character_short
```

```python
# Create empty dataframe to hold the results of data conversion
wand = pd.DataFrame()

# Change json single quotes to double quotes so that it can be normalized
character_short = character_short.replace({'\'': '"'}, regex=True)

# Loop through the data to fill the dataframe
for index in character_short.index:
    row_json = json.loads(character_short.wand[index])
    normalized_row = pd.json_normalize(row_json)
    wand = wand.append(normalized_row)

# Reset the index of the attributes dataframe
wand = wand.reset_index(drop=True)

# Drop the original attributes column
character_short = character_short.drop(columns=['wand'])

# Join the results
character_short = character_short.join(wand)

character_short.rename(columns={'name':'Name','species':'Species','gender':'Gender','house':'House',
                                'eyeColour':'Eye colour','hairColour':'Hair colour'},inplace=True)
character_short
```
